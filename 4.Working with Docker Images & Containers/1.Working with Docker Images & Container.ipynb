### A) Lifecycle

* `docker pull <image>` â€” **Download** image from a registry (e.g., Docker Hub).
* `docker run [opts] <image>` â€” **Create + start** a new container from an image.
* `docker ps` â€” **List running** containers.
* `docker ps -a` â€” **List all** containers (running + stopped).
* `docker stop <container>` â€” **Gracefully stop** a running container.
* `docker rm <container>` â€” **Remove** a **stopped** container.
* `docker start|restart <container>` â€” **Start/Restart** an existing container.

**Example**

```bash
docker pull nginx
docker run -d --name web nginx
docker ps
docker stop web && docker rm web
```

---

### B) Inspect & Interact

* `docker logs <container>` â€” **Show stdout/stderr** logs.
* `docker logs -f <container>` â€” **Follow** logs (tail -f).
* `docker exec -it <container> sh|bash` â€” **Open shell / run command** inside a running container.
* `docker inspect <name|id>` â€” **Detailed JSON** info (env, mounts, IP, etc.).
* `docker top <container>` â€” **Processes** running inside container.
* `docker stats [name]` â€” **Live CPU/RAM/IO** usage.

**Example**

```bash
docker exec -it web sh
docker logs -f web
docker inspect web | less
```

---

### C) Modes & Common Flags

* `-d` â€” **Detached** (background service/APIs).
* `-it` â€” **Interactive TTY** (shell/debug).
* `--name <name>` â€” **Friendly name** for the container.

**Examples**

```bash
docker run -it ubuntu sh     # interactive shell
docker run -d --name api myimg  # background service
```

---

### D) Networking & Ports

* `-p HOST:CONTAINER` â€” **Map host port â†’ container port**.
* `docker port <container>` â€” **Show** port mappings.
* `docker network create <net>` â€” **Create** user network.
* `docker run --network <net> ...` â€” **Attach** to network.

**Example**

```bash
docker run -d -p 8080:80 nginx     # visit http://localhost:8080
docker network create appnet
docker run -d --network appnet --name db postgres
```

---

### E) Storage (Volumes & Bind Mounts)

* `docker volume create <name>` â€” **Create** managed volume.
* `-v <volname>:/path/in/container` â€” **Mount volume** (persistent).
* `-v /host/path:/path/in/container` â€” **Bind mount** host folder.
* Add `:ro` to make **read-only**.

**Examples**

```bash
docker volume create appdata
docker run -d -v appdata:/var/lib/postgresql/data postgres

docker run -d -v "$(pwd)/site:/usr/share/nginx/html:ro" nginx
```

---

### F) Images (Build, Tag, Push, Prune)

* `docker images` â€” **List** images.
* `docker build -t <name>:tag .` â€” **Build** image from **Dockerfile** in current dir.
* `docker tag src dest` â€” **Retag** an image (e.g., for pushing).
* `docker push <repo>:tag` â€” **Upload** image to registry.
* `docker rmi <image>` â€” **Remove** image.
* `docker image prune` â€” **Remove dangling** (unused) layers.

**Examples**

```bash
docker build -t myapp:latest .
docker tag myapp:latest username/myapp:latest
docker push username/myapp:latest
```

---

### G) Copy Files (Host â†” Container)

* `docker cp SRC DEST` â€” **Copy files** in/out of containers.

**Examples**

```bash
docker cp web:/etc/nginx/nginx.conf ./nginx.conf   # container â†’ host
docker cp ./index.html web:/usr/share/nginx/html/  # host â†’ container
```

---

### H) Env & Restart Policies

* `-e KEY=VAL` â€” **Set env var** inside container.
* `--env-file .env` â€” **Load many env vars** from file.
* `--restart unless-stopped|always|on-failure` â€” **Auto-restart** policy.

**Example**

```bash
docker run -d --restart unless-stopped -e POSTGRES_PASSWORD=pass postgres
```

---

### I) Cleanup (Safe Defaults)

* `docker rm -f <container>` â€” **Stop + remove** container.
* `docker system prune -f` â€” **Remove unused** containers/images/networks (keeps volumes).
* `docker system prune -a -f` â€” Also removes **unused images** (be careful).
