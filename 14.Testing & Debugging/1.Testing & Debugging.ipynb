### üõ†Ô∏è **Debugging Inside Containers**

* **Interactive Shell Access:**

```bash
docker exec -it <container_name> /bin/sh
# or for Debian/Ubuntu-based images
docker exec -it <container_name> /bin/bash
```

* Use this to:

  * Inspect file paths
  * Debug environment variables
  * Manually run commands

---

### üß™ **Temporary Containers for Testing**

* **Create and test without running a full containerized app:**

```bash
docker run --rm -it alpine sh
```

* Use `--rm` to auto-remove after exit
* Helpful for testing base images, commands, configs, tools, etc.

---

### üìã **Logs & Stats**

| Command              | Description                               |
| -------------------- | ----------------------------------------- |
| `docker logs <name>` | Shows logs of a running/stopped container |
| `docker stats`       | Displays real-time CPU, memory, I/O usage |

* Add `-f` to `logs` to follow output in real-time
* Use `--no-stream` with `stats` for a single snapshot

```bash
docker logs -f my_app
docker stats --no-stream
```
