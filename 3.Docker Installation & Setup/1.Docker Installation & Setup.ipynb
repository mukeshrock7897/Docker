# 📦 2. Docker Installation & Setup

### 🧭 Choose Your Path

* **Docker Desktop (macOS/Windows)** → Best for dev machines. Includes Engine + CLI + Compose.
* **CLI-only (Linux/Servers)** → Lightweight, headless.

---

### 🖥️ Docker Desktop (macOS/Windows)

* Download & install **Docker Desktop**.
* Start the app once.
* You get `docker` + `docker compose` out of the box.

Quick check:

```bash
docker version
docker compose version
```

---

### 🐧 Linux (Ubuntu/Debian – minimal & safe)

```bash
sudo apt update
sudo apt install -y docker.io docker-compose-plugin
sudo systemctl enable --now docker

# allow running docker without sudo
sudo usermod -aG docker $USER
# re-login OR:
newgrp docker
```

---

### ✅ Verify Install

```bash
docker --version
docker info
docker run --rm hello-world
```

Expected: **“Hello from Docker!”**

---

### 🔐 Sign in (optional but useful)

```bash
docker login
# enter Docker Hub username & password
```

---

### 🆚 Desktop vs CLI (quick view)

| Feature  | Desktop (macOS/Win) | CLI-only (Linux)        |
| -------- | ------------------- | ----------------------- |
| GUI      | ✅                   | ❌                       |
| Best for | Dev/laptops         | Servers/CI              |
| Compose  | Built-in            | `docker-compose-plugin` |

---

### 🛠️ Common Fixes

```bash
# Permission denied → re-login or:
newgrp docker

# Service issues (Linux):
systemctl status docker
sudo journalctl -u docker --no-pager
```
