### ‚úÖ **Installing Docker on Windows / macOS / Linux**

**Option 1: Docker Desktop (Recommended for Beginners)**

* All-in-one UI + CLI package for Windows/macOS
* Includes Docker Engine, CLI, Docker Compose

üîó [Install Docker Desktop ‚Üí](https://www.docker.com/products/docker-desktop/)

**Option 2: CLI-only (for Linux Servers/Cloud VMs)**
Use the following commands:

```bash
# Ubuntu / Debian
sudo apt update
sudo apt install docker.io
sudo systemctl start docker
sudo systemctl enable docker
```

‚úÖ Add your user to the docker group:

```bash
sudo usermod -aG docker $USER
```

### üîç **Verifying Installation**

After installation:
