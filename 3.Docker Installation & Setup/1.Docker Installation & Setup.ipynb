# 📦 **2. Docker Installation & Setup**

---

### 🧭 Choose Your Path

* **🐳 Docker Desktop (macOS/Windows)** → All-in-one (Engine + CLI + Compose).
* **🐧 CLI-only (Linux/Servers)** → Lightweight, no GUI.

---

### 🖥️ Docker Desktop (macOS/Win)

* Download & install **Docker Desktop**.
* Start once → you get `docker` + `docker compose`.

```bash
docker version
docker compose version
```

---

### 🐧 Linux (Ubuntu/Debian)

```bash
sudo apt update
sudo apt install -y docker.io docker-compose-plugin
sudo systemctl enable --now docker

# Run docker without sudo
sudo usermod -aG docker $USER
newgrp docker
```

---

### ✅ Verify install

```bash
docker --version
docker info
docker run --rm hello-world
```

👉 Expect: **“Hello from Docker!”**

---

### 🔐 Sign in (optional)

```bash
docker login
```

Use your **Docker Hub** credentials.

---

### 🆚 Desktop vs CLI (quick view)

| Feature  | Desktop (Mac/Win) | CLI-only (Linux) |
| -------- | ----------------- | ---------------- |
| GUI      | ✅ Yes             | ❌ No             |
| Best for | Dev/laptops       | Servers/CI       |
| Compose  | Built-in          | Plugin needed    |

---

### 🛠️ Common Fixes

```bash
# Permission denied
newgrp docker

# Service issues (Linux)
systemctl status docker
sudo journalctl -u docker --no-pager
```
