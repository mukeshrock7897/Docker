## 🔄 **8. Docker Image Management**

---

### 🏷️ **Tagging Images**

Use `docker tag` to give your image a name (often for pushing):

```bash
docker tag myapp:latest myusername/myapp:v1
```

📌 **Format**:
`[registry]/[username]/[image-name]:[tag]`

---

### 📤 **Pushing to Docker Hub or Private Registry**

```bash
docker push myusername/myapp:v1
```

🔐 Make sure you're logged in:

```bash
docker login
```

You can also push to:

* Docker Hub (default)
* AWS ECR
* GitHub Container Registry
* Private registry (e.g., Harbor)

---

### 💾 **Saving & Loading Images**

**Save to a `.tar` file** (e.g., for backup or offline transfer):

```bash
docker save -o myapp.tar myapp:latest
```

**Load from file**:

```bash
docker load -i myapp.tar
```

---

### 🧹 **Cleaning Up Unused Images, Containers, Volumes**

| Command                  | What It Does                                        |
| ------------------------ | --------------------------------------------------- |
| `docker image prune`     | Removes unused images                               |
| `docker container prune` | Removes stopped containers                          |
| `docker volume prune`    | Removes unused volumes                              |
| `docker system prune`    | Removes all unused (images + containers + networks) |

🧪 Example:

```bash
docker system prune -a  # Warning: removes everything unused!
```
